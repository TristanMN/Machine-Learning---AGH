```python
#1 K-means clustering - example I

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_blobs

#2

X, y = make_blobs(n_samples = 500, n_features = 2, centers = 4, cluster_std=0.60)

#3

print(X.shape, y.shape)
```

```python
#4

plt.scatter(X[:,0],X[:,1], c=y)
plt.show()
```

```python
#5

from sklearn.cluster import KMeans

kmeans = KMeans(init='random', n_clusters=4, random_state=13, n_init="auto").fit(X)
y_pred = kmeans.predict(X)

#6
plt.scatter(X[:,0],X[:,1], c=y_pred)
```

```python
#7

inertia = kmeans.inertia_
print("Inertia:", inertia)
```

```python
#8

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

algorithms = ["single","average","complete","ward"]
X = np.array(X)
plt.figure(figsize = (16,20))
for n in range(5,10):
    for m in range(4):
        algorithm = AgglomerativeClustering(n_clusters=n, linkage=algorithms[m]).fit(X)
        y_pred = algorithm.fit_predict(X)
        plt.subplot(5, 4, (n-5)*4+m+1)
        title = str(algorithms[m]) + " " + str(n) + " clusters"
        plt.title(title)
        plt.scatter(X[:, 0], X[:, 1], c=y_pred)
        plt.axis('off')
```

```python
#1 Selecting the number of clusters with silhouette analysis on KMeans clustering

X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)

#2

cluster_range = np.arange(2,6)

#3

sc = []
for n in cluster_range:
    kmeans = KMeans(n_clusters = n, random_state = 13).fit(X)
    y_pred = kmeans.fit_predict(X)
    score = silhouette_score(X, y_pred)
    sc.append(score)
    print(score, " for ", n, " clusters")
plt.plot(cluster_range,sc)
```

```python
#DBSCAN - the best scan

from sklearn.cluster import DBSCAN

X, y = make_moons(n_samples=200, noise=.05, random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(X)
labels = db.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```
